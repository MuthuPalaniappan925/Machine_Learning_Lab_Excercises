### Importing Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

### Generating Random Points

```python
np.random.seed(9) #seed
```

```python
#generating x points
x = np.random.random(size=(1000)) #generating 1000 random points
m=c=1
y = m*x+c+np.random.normal(loc=0,scale=0.2,size=1000)
```

### Visualizing the datapoints X and Y

```python
sns.scatterplot(x,y,color='red');
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Model Building

```python
m = 0
c = 0
#learning rate
L = 0.0001
#number of iterations to perform gradient descent
epochs = 100000
#number of datapoints
n = float(len(x))
#performing gradient decesent
for i in range(epochs):
    y_pred = m*x+c
    #der wrt to m
    d_m = (-2/n) * sum(x*(y-y_pred))
    #der wrt to c
    d_c = (-2/n) * sum(y-y_pred)
    #updating m and c
    m = m-L*d_m
    c = c-L*d_c
#printing m and c
print(f"Slope M : {m}\nIntercept C : {c}")
```

### Prediction

```python
y_pred = m*x+c
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
plt.show()
```

### Comparing with sklearn Package

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x.reshape(-1,1),y)
```

```python
sk_intercept = LR.intercept_
sk_coef = LR.coef_[0]
print(f"Intercept C : {sk_intercept}\nSlope M : {sk_coef}")
```
